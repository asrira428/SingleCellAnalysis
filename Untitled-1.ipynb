```python
path = "GSE72056_melanoma_single_cell_revised_v2.txt"

df = pd.read_csv(path, sep="\t", index_col=0)

# take a peek at the top
df
```

```python
df_t = df.T
df_t
```

```python
# Assume scrna is your DataFrame with genes as rows and cell-IDs as columns
import pandas as pd

# Extract patient name for each column (prefix before first underscore)
patient_names = df.columns.to_series().str.split('_').str[0]

# Group the column names by patient
cols_by_patient = (
    pd.DataFrame({'cell_id': df.columns, 'patient': patient_names})
      .groupby('patient')['cell_id']
      .apply(list)
      .to_dict()
)

# Inspect number of cells per patient
for patient, cells in cols_by_patient.items():
    print(f"{patient}: {len(cells)} cells")
    print(cells[:5], '...')  # Show first 5 cell IDs per patient
    print()
```

```python
# Get top 2 patients by number of cells
top_patients = sorted(cols_by_patient.items(), key=lambda x: len(x[1]), reverse=True)[:5]

# Display results
for patient, cells in top_patients:
    print(f"{patient}: {len(cells)} cells")
    print(cells[:5], '...')  # Show first 5 cell IDs
    print()
```

```python
# Subset rows (cells) belonging to patient CY89A
df_cy89a = df_t[df_t.index.str.startswith("Cy81")]

# Inspect result
print(df_cy89a.shape)
df_cy89a['malignant(1=no,2=yes,0=unresolved)'].value_counts()
```

```python
df_cy89a
```

```python
from scipy.stats import ttest_ind
import pandas as pd

# Filter out unresolved rows
df_valid = df_cy89a[df_cy89a["malignant(1=no,2=yes,0=unresolved)"] != 0]

# Split into malignant and non-malignant
malignant = df_valid[df_valid["malignant(1=no,2=yes,0=unresolved)"] == 2]
non_malignant = df_valid[df_valid["malignant(1=no,2=yes,0=unresolved)"] == 1]

# Extract gene expression columns
gene_expr_cols = df_cy89a.columns[3:]

pvals = []

# Loop through genes and compute t-test only if both groups have variability
for gene in gene_expr_cols:
    try:
        mal_values = pd.to_numeric(malignant[gene], errors='coerce')
        nonmal_values = pd.to_numeric(non_malignant[gene], errors='coerce')

        if mal_values.nunique() > 1 and nonmal_values.nunique() > 1:
            stat, p = ttest_ind(mal_values.dropna(), nonmal_values.dropna(), equal_var=False)
            if isinstance(p, float):  # make sure p is a scalar
                pvals.append((gene, p))
    except:
        continue  # skip genes with issues

# Create and sort p-value DataFrame
deg_df = pd.DataFrame(pvals, columns=["gene", "pval"])
deg_df = deg_df.sort_values("pval")

# Get top 500 genes
top_500_genes = deg_df["gene"].iloc[:500].tolist()

# Combine metadata with those top genes
metadata_cols = df_cy89a.columns[:3]
df_top500 = pd.concat([df_cy89a[metadata_cols], df_cy89a[top_500_genes]], axis=1)
```

```python
df_top500
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, confusion_matrix
import pandas as pd

# Extract gene expression data (exclude first 3 metadata columns)
X = df_top500.iloc[:, 3:].values  # shape: (n_samples, 500)

# Perform k-means clustering (2 clusters: malignant vs non-malignant)
kmeans = KMeans(n_clusters=2, random_state=42, n_init='auto')
df_top500['cluster'] = kmeans.fit_predict(X)

# Get true labels (1 = non-malignant, 2 = malignant)
true_labels = df_top500["malignant(1=no,2=yes,0=unresolved)"].values

# Filter out unresolved rows (label == 0)
valid_mask = (true_labels != 0)
true_labels = true_labels[valid_mask]
predicted_clusters = df_top500['cluster'].values[valid_mask]

# Compare clusters to true labels
from sklearn.metrics import classification_report

print("Adjusted Rand Index:", adjusted_rand_score(true_labels, predicted_clusters))

# Optional: Map cluster labels to match majority class
pd.crosstab(true_labels, predicted_clusters, rownames=['True'], colnames=['Predicted'])
```

```python
df_top500 = df_top500.rename(columns={
    'malignant(1=no,2=yes,0=unresolved)': 'malignant',
    'non-malignant cell type (1=T,2=B,3=Macro.4=Endo.,5=CAF;6=NK)': 'non_malignant_cell_type'
})
```

```python
import matplotlib.pyplot as plt

# Step 1: Get gene expression matrix (from column 4 onward)
gene_expr = df_top500.iloc[:, 4:]

# Step 2: Sum expression per cell
total_expr = gene_expr.sum(axis=1)

# Step 3: Assign colors based on malignancy
colors = ['red' if val == 1 else 'blue' for val in df_top500['malignant']]

# Step 4: Plot
plt.figure(figsize=(16, 5))
plt.bar(range(len(total_expr)), total_expr, color=colors, edgecolor='black')

# Beautify
plt.xlabel("Cell Index")
plt.ylabel("Total Gene Expression")
plt.title("Total Gene Expression per Cell (Red = Non-malignant, Blue = Malignant)")
plt.tight_layout()
plt.show()
```
