```python
import pandas as pd
```

```python
path = "GSE72056_melanoma_single_cell_revised_v2.txt"

df = pd.read_csv(path, sep="\t", index_col=0)

# take a peek at the top
df
```

```python
scrna = df.iloc[3:, :]  
scrna
```

```python
df_t = df.T
df_t
```

```python
# Assume scrna is your DataFrame with genes as rows and cell-IDs as columns
import pandas as pd

# Extract patient name for each column (prefix before first underscore)
patient_names = scrna.columns.to_series().str.split('_').str[0]

# Group the column names by patient
cols_by_patient = (
    pd.DataFrame({'cell_id': scrna.columns, 'patient': patient_names})
      .groupby('patient')['cell_id']
      .apply(list)
      .to_dict()
)

# Inspect number of cells per patient
for patient, cells in cols_by_patient.items():
    print(f"{patient}: {len(cells)} cells")
    print(cells[:5], '...')  # Show first 5 cell IDs per patient
    print()
```

```python
# Get top 2 patients by number of cells
top_patients = sorted(cols_by_patient.items(), key=lambda x: len(x[1]), reverse=True)[:5]

# Display results
for patient, cells in top_patients:
    print(f"{patient}: {len(cells)} cells")
    print(cells[:5], '...')  # Show first 5 cell IDs
    print()
```

```python
# Subset rows (cells) belonging to patient CY89A
df_cy89a = df_t[df_t.index.str.startswith("Cy81")]

# Inspect result
print(df_cy89a.shape)
df_cy89a['malignant(1=no,2=yes,0=unresolved)'].value_counts()
```

```python
df_cy89a['malignant(1=no,2=yes,0=unresolved)'].value_counts()
```

```python
metadata_cols = df_cy89a.columns[:3]

# Compute column-wise sum for gene expression columns (from column 4 onward)
gene_expr = df_cy89a.iloc[:, 4:]
top_250_genes = gene_expr.sum(axis=0).nlargest(250).index

# Combine metadata with top genes
df_top250 = pd.concat([df_cy89a[metadata_cols], gene_expr[top_250_genes]], axis=1)
df_top250
```

```python
df_top250 = df_top250.rename(columns={
    'malignant(1=no,2=yes,0=unresolved)': 'malignant',
    'non-malignant cell type (1=T,2=B,3=Macro.4=Endo.,5=CAF;6=NK)': 'non_malignant_cell_type'
})
```

```python
df_top250
```

```python
import matplotlib.pyplot as plt

# Step 1: Get gene expression matrix (from column 4 onward)
gene_expr = df_top250.iloc[:, 4:]

# Step 2: Sum expression per cell
total_expr = gene_expr.sum(axis=1)

# Step 3: Assign colors based on malignancy
colors = ['red' if val == 1 else 'blue' for val in df_top250['malignant']]

# Step 4: Plot
plt.figure(figsize=(16, 5))
plt.bar(range(len(total_expr)), total_expr, color=colors, edgecolor='black')

# Beautify
plt.xlabel("Cell Index")
plt.ylabel("Total Gene Expression")
plt.title("Total Gene Expression per Cell (Red = Non-malignant, Blue = Malignant)")
plt.tight_layout()
plt.show()
```
